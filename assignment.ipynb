# HEALTH DATASET

```python
import pandas as pd
import seaborn  as sns
import matplotlib.pyplot as plt
```

```python
# Load dataset
def load_dataset(file_path):
    return pd.read_csv(file_path)

df = load_dataset("./Health_heart_experimental.csv")
df.head(5)
```

## Line Chart
#### Variation of Heart Rate (HR) with age

```python
sns.lineplot(data=df, x='age', y='HR',color="blue", marker='o')

# add a title
plt.title("Heart Rate (HR) vs age")

plt.ylabel("Heart Rate")
plt.xlabel("Age")
plt.show()
```

#### The above graph shows that the Heart Rate (HR) does not have a strictly linear relationship with age, younger individuals may exhibit a broader range of heart rates due to activity levels or natural variations.

## Bar Plot

### SysBP and DiaBP by sex


```python
sysbp_data = df.groupby('sex')['SysBP'].mean()
ax = sysbp_data.plot(kind='bar', color=['blue', 'red'], alpha=0.7, figsize=(10, 6))

# Add chart title and labels
plt.title("Average Systolic Blood Pressure by Sex")
plt.xlabel("Sex")
plt.ylabel("Average SysBP (mmHg)")

# Improve x-axis ticks
plt.xticks(ticks=[0, 1], labels=["Female", "Male"], rotation=0)

# Annotate each bar with its corresponding value
for i, v in enumerate(sysbp_data):
    ax.text(i, v + 1, f'{v:.2f}', ha='center', va='bottom', fontsize=10)

# Show the plot
plt.show()

# Group data by 'sex' and calculate mean DiaBP
diabp_data = df.groupby('sex')['DiaBP'].mean()
ax = diabp_data.plot(kind='bar', color=['blue', 'red'], alpha=0.7, figsize=(10, 6))

# Add chart title and labels
plt.title("Average Diastolic Blood Pressure by Sex")
plt.xlabel("Sex")
plt.ylabel("Average DiaBP (mmHg)")

# Improve x-axis ticks
plt.xticks(ticks=[0, 1], labels=["Female", "Male"], rotation=0)

# Annotate each bar with its corresponding value
for i, v in enumerate(diabp_data):
    ax.text(i, v + 1, f'{v:.2f}', ha='center', va='bottom', fontsize=10)

# Show the plot
plt.show()
```

#### The above Bar Plot shows that the average SysBP and DiaBP of female and male are almost equal, which signifiesn males and females have relatively similar health conditions, particularly in terms of blood pressure. This could be due to lifestyle, diet, exercise, or genetic factors in this specific population.

## Pie Chart

### Distribution of Male and Female

```python
plt.figure(figsize=(8, 8))
df['sex'].value_counts().plot.pie(autopct='%1.1f%%', colors=['skyblue', 'salmon'], startangle=90)
plt.title("Distribution of Male and Female")
plt.xlabel(" (0=Female, 1=Male)")
plt.show()
```

#### The above chart shows that the population of female is slightly greater than male.

## Box Plot

```python
sns.boxplot(x='sex', y='BMI', data=df)
plt.xlabel("Sex (0=Female, 1=Male)")
plt.ylabel("BMI")
plt.show()
```

The above box plot shows that the BMI of both are similar.

## Histogram


### Frequency distribution of age

```python
plt.figure(figsize=(10, 6))
sns.histplot(df['age'], bins=20, kde=True, color='blue', alpha=0.8)
plt.title("Distribution of Age")
plt.xlabel("Age")
plt.ylabel("Frequency")
plt.show()
```

#### The above graph shows that the people of age 20 and 60 are large in frequency.

## Scatter Plot

### Correlation between SysBP and DiaBP.

```python
import seaborn as sns
import matplotlib.pyplot as plt

sns.scatterplot(x='SysBP', y='DiaBP', hue='sex', style='indication', data=df)
plt.xlabel("Systolic Blood Pressure (SysBP)")
plt.ylabel("Diastolic Blood Pressure (DiaBP)")
plt.show()
```
